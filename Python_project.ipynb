```python
import pandas as pd

Data = {
    'Product': ["Laptop", "Mobile", "Tablet", "None", "Mobile"],
    'Sales': [50000, 75000, 85000, 90000, 60000]
}

df = pd.DataFrame(Data)

print(df.isnull().sum())

df = df.dropna()

df = df.drop_duplicates()

print("Cleaned Data")
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

Data = {
    'Product': ["Laptop", "Mobile", "Tablet", "Headphone", "Mobile", "Powerbank"],
    'Sales': [50000, 75000, 25000, 90000, 60000, 30000]
}

df = pd.DataFrame(Data)

print("Top Products:")

plt.figure(figsize=(7,5))

plt.bar(df['Product'], df['Sales'])

plt.title("Top-Selling Products")

plt.xlabel("Product")

plt.ylabel("Sales")

plt.grid(axis='y')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month":["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug"],
    "Sales": [100000, 140000, 180000, 200000, 220000, 240000, 250000, 300000]
}

df = pd.DataFrame(data)

print("Monthly Sales data:")
print(df)

plt.figure(figsize=(8,5))

plt.plot(df["Month"], df["Sales"], marker='o')

plt.title("Monthly Sales Trends")

plt.xlabel("Month")

plt.ylabel("Sales Amount")

plt.grid(axis='y')

plt.show()
```
